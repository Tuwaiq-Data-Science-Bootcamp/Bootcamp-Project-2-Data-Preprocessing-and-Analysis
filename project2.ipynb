## Data Visualization

```python
#Confirmed_cases_per_Region

#Confirmed_cases= df.groupby(by='WHO Region').max()[['Confirmed']].sort_values(by ='Confirmed',ascending =False).reset_index()
Confirmed_cases= df.groupby(by='Region')['Confirmed'].sum().sort_values(ascending =False)

Confirmed_cases.plot.pie(title='Confirmed cases',colors = palette);
#plt.legend(['Confirmed cases','Region'], shadow = True, loc = 0);
plt.xlabel('Region')
plt.ylabel('Confirmed cases')
```

```python
# top 10 actave cases 
top_10_active= df.groupby(by='Region').max()[['Active']].sort_values(by = 'Active',ascending =False).reset_index()
fig= plt.figure(figsize=(20,10))
plt.title('top 5 actave cases in 2020 per Region',color='black',size=25)
sns.barplot(data= top_10_active.iloc[:5],y='Active', x='Region',hue="Active",linewidth=2, edgecolor='black')
plt.xlabel('Region')
plt.ylabel('Active cases')
plt.grid()
plt.show()
```

```python
# top Region with highest deaths
top_10_active= df.groupby(by='Region').max()[['Deaths']].sort_values(by = 'Deaths',ascending =False).reset_index()
fig= plt.figure(figsize=(20,10))
plt.title('top Region with highest deaths',color='black',size=14)
sns.barplot(data= top_10_active.iloc[:5],y='Deaths', x='Region',linewidth=1, edgecolor='black',hue="Deaths",palette='coolwarm')
plt.xlabel('Region')
plt.ylabel('Death cases')
plt.show()
```

```python
#The rate of recovery by Region
#df2= df.groupby('WHO Region')['Recovered'].sum().sort_values(ascending =False)
plt.figure(figsize=(20,10))
sns.kdeplot(data=df,x='Recovered',hue='Region',palette='coolwarm')
#plt.legend(['New cases','Country'], shadow = True, loc = 0);
plt.title('The rate of recovery by Region ', size=14)

```

```python
top_actives1=df1.groupby("country")["Deaths"].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(12,4)) 

sns.swarmplot(
    x="country",
    y="Deaths",  
    data=top_actives1,
    hue="Deaths",
    palette='coolwarm'
)
plt.title('top 10 Country with highest Deaths cases ', size=14);
```

```python
top_actives2=df1.groupby("country")["Recovered"].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(15,4))
plt.plot(top_actives2['country'],top_actives2['Recovered'],color= palette[0]);
plt.legend(['Recovered','Country'], shadow = True, loc = 0);
plt.xlabel('Country')
plt.ylabel('Recovered cases')
plt.title('top 10 countries with highest Recovered cases', size=14)
```

```python
top_actives3=df1.groupby("country")["Active"].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(20,5))
plt.bar(top_actives3['country'],top_actives3['Active'],color = palette);
#plt.legend(['Active','Country'], shadow = True, loc = 0);
plt.xlabel('Country')
plt.ylabel('Active cases')
plt.title('top 10 countries with highest Active cases', size=14);
```
