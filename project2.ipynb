```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("country_wise_latest.csv")
df
```

```python
#Confirmed_cases_per_Region

#Confirmed_cases= df.groupby(by='WHO Region').max()[['Confirmed']].sort_values(by ='Confirmed',ascending =False).reset_index()
Confirmed_cases= df.groupby(by='WHO Region')['Confirmed'].sum().sort_values(ascending =False)
plt.figure(figsize=(20,10))
Confirmed_cases.plot.pie(title='Confirmed cases');
```

```python
# top 10 actave cases 
top_10_active= df.groupby(by='WHO Region').max()[['Active']].sort_values(by = 'Active',ascending =False).reset_index()
fig= plt.figure(figsize=(20,10))
plt.title('top 5 actave cases in 2020 per Region',color='black',size=25)
ax=sns.barplot(data= top_10_active.iloc[:5],y='Active', x='WHO Region',linewidth=2, edgecolor='black')
plt.xlabel('Region')
plt.ylabel('Active cases')
plt.grid()
plt.show()
```

```python
# top Region with highest deaths
top_10_active= df.groupby(by='WHO Region').max()[['Deaths']].sort_values(by = 'Deaths',ascending =False).reset_index()
fig= plt.figure(figsize=(20,10))
plt.title('top Region with highest deaths',color='black',size=25)
ax=sns.barplot(data= top_10_active.iloc[:5],y='Deaths', x='WHO Region',linewidth=2, edgecolor='black')
plt.xlabel('Region')
plt.ylabel('Death cases')
plt.show()
```

```python
#The rate of recovery by Region
#df2= df.groupby('WHO Region')['Recovered'].sum().sort_values(ascending =False)
plt.figure(figsize=(20,10))
sns.kdeplot(
     data=df, 
     x='Recovered',
     hue='WHO Region')

plt.title("The rate of recovery by Region",fontsize = 20, loc='center', color='black' );
```
