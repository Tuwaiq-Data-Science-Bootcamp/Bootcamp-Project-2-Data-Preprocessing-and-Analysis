## ~FIlling The Non Values

```python
#Replace NULL values in mutiple columns with 'Unknown'
Jobs.Gender=Jobs.Gender.fillna('Unknown')
Jobs['Job Location']=Jobs['Job Location'].fillna('Unknown')
Jobs['Residence Location']=Jobs['Residence Location'].fillna('Unknown')
Jobs['Career Level']=Jobs['Career Level'].fillna('Unknown')
Jobs['Degree']=Jobs['Degree'].fillna('Unknown')
Jobs['Degree']=Jobs['Degree'].fillna('Unknown')
Jobs['Job Country']=Jobs['Job Country'].fillna('Unknown')
Jobs['Job City']=Jobs['Job City'].fillna('Unknown')

#Replace NULL values in mutiple columns with the mean
Jobs["Min Years of Experience"]=Jobs['Min Years of Experience'].fillna(Jobs['Min Years of Experience'].mean())
Jobs["Max Years of Experience"]=Jobs['Max Years of Experience'].fillna(Jobs['Max Years of Experience'].mean())
Jobs["Monthly Salary Min Range"]=Jobs['Monthly Salary Min Range'].fillna(Jobs['Monthly Salary Min Range'].mean())
Jobs["Monthly Salary Max Range"]=Jobs['Monthly Salary Max Range'].fillna(Jobs['Monthly Salary Max Range'].mean())
Jobs["Min Age"]=Jobs['Min Age'].fillna(Jobs['Min Age'].mean())
Jobs["Max Age"]=Jobs['Max Age'].fillna(Jobs['Max Age'].mean())

Jobs.isnull().sum()
```

```python
sns.set_style("dark");
sns.set_palette("Paired")
```

```python
plt.figure(figsize=(50,8))
sns.histplot(Jobs["Job City"]) 
plt.title("Jobs by City",fontsize = 30, loc='center', color='k');
```

```python
roles = Jobs['Job Role'].unique()
role_count = Jobs.value_counts('Job Role')
plt.figure(figsize=(7,7))
plt.pie(role_count, labels=roles);
plt.title('Jobs by Role', fontsize = 30, loc='center', color='k');
```

```python
sns.set(font_scale=1)
sns.set(rc={"figure.figsize":(15, 10)})
sns.countplot( y='Job City',data=Jobs).set_title('Number of Jobs by cities');
```

```python
plt.figure(figsize=(14,5));# it does not scale well to large numbers of observations.

sns.swarmplot(
    x='Min Years of Experience',
    y="Monthly Salary Max Range", 
    hue='Job Role', 
    data=Jobs2
);

plt.title("The Jobs Salaries Based on the Experience");
plt.xlabel('Years of Experience');
plt.ylabel('Highest Monthly Salary');
plt.figure(figsize=(14,5));
```
