```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('SA_Aqar.csv')
```

```python
# removes details column
df.drop('details', inplace = True, axis = 1)
```

```python
# print duplicated values
df.duplicated().sum()
```

```python
# removes duplicated values
df.drop_duplicates(inplace= True)
```

```python
df.shape
```

```python
# removes outliers from size column
cols = ['size'] 

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

# Data Visualization

```python
# plot1: nada
# Question 1: Is the data set balanced?
plt.figure(figsize=(10,5))
plt.title('Question 1: Is the data set balanced?', fontsize=20)
df.city.value_counts().plot(kind='barh', color ='#0E5E6F');
plt.xlabel('Count', fontsize=15)
plt.ylabel('City', fontsize=15)
plt.show()
```

```python
# plot2: nada
# Question 2: Which city is the most expensive?
plt.figure(figsize=(10,5))
plt.title('Question 2: Which city is the most expensive?', fontsize=20)
df.groupby(df['city']).mean()['price'].plot(kind='barh', color ='#0E5E6F');
plt.xlabel('City', fontsize = 15)
plt.ylabel('Price (Mean)', fontsize = 15)
plt.show();
```

```python
# plot3: nada
# Question 3: is there a relationship between property_age and price?

df_riyadh = df[(df['city'] == 'Riyadh')]
df_khobar = df[(df['city'] == 'Khobar')]
df_dammam = df[(df['city'] == 'Dammmam')]
df_jeddah = df[(df['city'] == 'Jeddah')]

plt.figure(figsize=(20,15))
plt.subplot(2, 2, 1)
plt.title('Riyadh', fontsize=20)
sns.lineplot(x = df_riyadh.property_age, y = df_riyadh.price);
plt.axvline(x = 32, color = 'r')
plt.xlabel('property_age', fontsize = 15)
plt.ylabel('Price', fontsize = 15)


plt.subplot(2, 2, 2)
plt.title('Jeddah', fontsize = 20)
sns.lineplot(x = df_jeddah.property_age, y = df_jeddah.price);
plt.axvline(x = 23, color = 'r')
plt.xlabel('property_age', fontsize = 15)
plt.ylabel('Price', fontsize = 15)

plt.subplot(2, 2, 3)
plt.title('Khobar', fontsize = 20)
sns.lineplot(x = df_khobar.property_age, y = df_khobar.price);
plt.axvline(x = 12, color = 'r')
plt.xlabel('property_age', fontsize = 15)
plt.ylabel('Price', fontsize = 15)

plt.subplot(2, 2, 4)
plt.title('Dammam', fontsize = 20)
sns.lineplot(x = df_dammam.property_age, y = df_dammam.price);
plt.axvline(x = 11, color = 'r')
plt.xlabel('property_age', fontsize = 15)
plt.ylabel('Price', fontsize = 15)

plt.suptitle('Question 2: What is the relationship between property age and price?', fontsize=25)
plt.show()
```

```python
# plot4: nada
# Question 4: How many bedrooms do most rental homes over $100,000 have?
plt.figure(figsize=(10,5))
plt.title('Question 4: How many bedrooms do most homes have?',fontsize = 20)
df5 = df[(df['price'] >= 100000)]
sns.countplot(x = df5['bedrooms'],palette='crest')
plt.xlabel('Bedrooms', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

```python
# plot5: nada
# Question 5: What is the column that affects the price column the most?
plt.subplots(figsize=(30,30))
p = sns.heatmap(df.corr(), annot=True, lw=1.5, fmt='.2f', cmap='coolwarm', vmin= -1, vmax=1)
rotxlabel = p.set_xticklabels(p.get_xticklabels(),fontdict={'fontsize':20}, rotation=90)
rotylabel = p.set_yticklabels(p.get_yticklabels(),fontdict={'fontsize':20}, rotation=30)
```

```python
# plot6: dalal
plt.figure(figsize=(10,5))
plt.title('Question 6: Where are the largest properties/houses?',fontsize = 20)
df3 = df.groupby(['city']).mean()['size']
df3.plot(kind='bar', color = '#0E5E6F')
plt.xlabel('City')
plt.ylabel('Area sqm')
plt.show()
```

```python
# plot7: dalal and nada
# the impact of fronts on the price 
df_riyadh = df[(df['city'] == 'Riyadh')]
df_khobar = df[(df['city'] == 'Khobar')]
df_dammam = df[(df['city'] == 'Dammmam')]
df_jeddah = df[(df['city'] == 'Jeddah')]

plt.figure(figsize=(20,15))
plt.title('Question 7: How does the building front affict the rental price?',fontsize = 20)

# first plot
plt.subplot(2, 2, 1)
plt.title('Riyadh', fontsize=20)

riyadh = df_riyadh.groupby(df_riyadh['front']).mean()['price']
riyadh_label = ['3 fronts', '4 fronts', 'South', 'South East',
'South West', 'East', 'North', 'North East', 'North West', 'West']
myexplode = [0, 0.3, 0, 0, 0, 0, 0, 0, 0, 0]
colors = sns.color_palette('pastel')[0:10]
plt.pie(riyadh, labels = riyadh_label, explode = myexplode,colors = colors, autopct='%1.1f%%')

# second plot
plt.subplot(2, 2, 2)
plt.title('Jeddah', fontsize=20)

Jeddah = df_jeddah.groupby(df_jeddah['front']).mean()['price']
Jeddah_label = ['3 fronts', '4 fronts', 'South', 'South East',
'South West', 'East', 'North', 'North East', 'North West', 'West']
colors = sns.color_palette('pastel')[0:10]
myexplode = [0.3, 0, 0, 0, 0, 0, 0, 0, 0, 0]
plt.pie(Jeddah,labels = Jeddah_label,explode = myexplode,colors = colors, autopct='%1.1f%%' )


# third plot
plt.subplot(2, 2, 3)
plt.title('Dammam', fontsize=20)

dammam = df_dammam.groupby(df_dammam['front']).mean()['price']
dammam_label = ['South', 'South East','East', 'North', 'North East', 'North West', 'West']
myexplode = [0, 0, 0, 0, 0.3, 0, 0]
colors = sns.color_palette('pastel')[0:7]
plt.pie(dammam,labels=dammam_label,explode = myexplode,colors = colors,  autopct='%1.1f%%')

# fourth plot
plt.subplot(2, 2, 4)
plt.title('Khobar', fontsize=20)
khobar = df_khobar.groupby(df_khobar['front']).mean()['price']
khobar_label = ['3 fronts','South', 'South East','South West', 'East', 'North', 'North East', 'North West', 'West']
myexplode = [0, 0, 0.3, 0, 0, 0, 0,0,0]
colors = sns.color_palette('pastel')[0:9]
plt.pie(khobar,labels = khobar_label,explode = myexplode,colors = colors, autopct='%1.1f%%' )
plt.show()
```

```python
# plot8: dalal
#The impact of having a driver room on the price
plt.figure(figsize=(10,5))
plt.title('Question 8: What is the affict of driver room availability on the price?',fontsize = 20)
sns.barplot(data=df, x="city",y="price", hue = "driver_room",ci=None, color ='#0E5E6F');
plt.xlabel('City')
plt.ylabel('price')
plt.show()
```

```python
# plot9: dalal
plt.figure(figsize=(10,5))
plt.title('Question 9: How does frontyard increas the value of your property ?',fontsize = 20)
sns.barplot(data=df,x="city",y="price", hue = "frontyard",ci=None, color ='#0E5E6F')
plt.xlabel('City')
plt.ylabel('price')
plt.show()
```

```python
# plot10: dalal
plt.figure(figsize=(15,15))
plt.title('Question 10: Does the availability of a swimming pool in the house affect the rental price?',fontsize = 20)
df5 = df[(df['price'] >= 100000)]
sns.swarmplot(data=df5, x='city', y='price', hue='pool', s = 10, alpha = 0.6);
plt.xlabel('city', fontsize = 15)
plt.ylabel('price', fontsize = 15)
plt.legend(loc = 'upper right', fontsize=15);
plt.ylim(50000, 850000)
```
